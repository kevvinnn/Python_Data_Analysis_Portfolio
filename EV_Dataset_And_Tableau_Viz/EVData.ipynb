**Data Preparation**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Electric_Vehicle_Population_Data.csv')
```

```python
df['Vehicle Location'] = df['Vehicle Location'].str.replace('POINT', '')
```

```python
df['LAT'] = df['Vehicle Location'].str.split(' ', expand=True)[1]
df['LON'] = df['Vehicle Location'].str.split(' ', expand=True)[2]
```

```python
df = df[[
    'VIN (1-10)', 'County', 'City', 'State', 'Postal Code', 'Model Year',
       'Make', 'Model', 'Electric Vehicle Type',
       'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 'Electric Range',
      # 'Base MSRP',  'Legislative District','DOL Vehicle ID','Vehicle Location',
     'Electric Utility', '2020 Census Tract', 'LAT',
       'LON'
]]
```
