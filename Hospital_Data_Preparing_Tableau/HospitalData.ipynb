```python
encounters = encounters[[
     #  'Id',
       'START', 'STOP',
       #'PATIENT', 'ORGANIZATION',  'PAYER',
       'ENCOUNTER_CLASS',
    # 'CODE',
     'DESCRIPTION', 'BASE_ENCOUNTER_COST',
       'TOTAL_CLAIM_COST', 'PAYER_COVERAGE',
     #'REASON_CODE',
       'REASON_DESCRIPTION', 'PAYER_NAME',
     # 'STATE_HEADQUARTERED',
       #'PHONE',
       'BIRTHDATE', 'DEATHDATE',
       #'PREFIX',
       'FIRST', 'LAST',
       #'SUFFIX', 'MAIDEN',
       'MARITAL', 'RACE', 'ETHNICITY', 'GENDER', 'BIRTHPLACE',
      'ADDRESS','CITY', 'STATE', 'COUNTY', 'ZIP', 'LATITUDE', 'LONGITUDE'
]]
```

```python
encounters['START'] = pd.to_datetime(encounters['START'])
encounters['STOP'] = pd.to_datetime(encounters['STOP'])

encounters['BIRTHDATE'] = pd.to_datetime(encounters['BIRTHDATE'])
encounters['DEATHDATE'] = pd.to_datetime(encounters['DEATHDATE'])

encounters['AGE'] = encounters['START'].dt.year - encounters['BIRTHDATE'].dt.year

encounters['DURATION_IN_MINS'] = (encounters['STOP'] - encounters['START']).dt.total_seconds() / 60

encounters['FIRST'] = encounters['FIRST'].apply(lambda value: ''.join([i for i in value if not i.isdigit()]))
encounters['LAST'] = encounters['LAST'].apply(lambda value: ''.join([i for i in value if not i.isdigit()]))


encounters['PATIENT_NAME'] = encounters['FIRST'] + ' ' + encounters['LAST']
```
